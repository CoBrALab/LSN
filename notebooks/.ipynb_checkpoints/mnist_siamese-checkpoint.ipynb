```python
# Package imports
%matplotlib inline
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random
import time
import math
from datetime import datetime
```

```python
class siamese_net(object):
  
    def __init__(self, layer_config, tune_config):
              
      self.input_L = tf.placeholder(tf.float32, [None, layer_config['input_shape']])
      self.input_R = tf.placeholder(tf.float32, [None, layer_config['input_shape']])
      self.labels = tf.placeholder(tf.float32, [None,1])   
      self.batch_size = tune_config['batch_size']
      self.dropout_f = 1

      with tf.variable_scope("siamese") as scope:
          self.branch_L = self.mlpnet(self.input_L)
          scope.reuse_variables()
          self.branch_R = self.mlpnet(self.input_R)
          
      # Create loss      
      self.distance = tf.sqrt(tf.reduce_sum(tf.pow(tf.subtract(self.branch_L,self.branch_R),2),1,keepdims=True))
      self.loss = self.contrastive_loss()

    #MLP
    def mlpnet(self, X):
      l1 = self.mlp(X,layer_config['input_shape'],layer_config['l1'],name='l1')
      l1 = tf.nn.dropout(l1,self.dropout_f)
      l2 = self.mlp(l1,layer_config['l1'],layer_config['l2'],name='l2')
      l2 = tf.nn.dropout(l2,self.dropout_f)
      l3 = self.mlp(l2,layer_config['l2'],layer_config['l3'],name='l3')
      l3 = tf.nn.dropout(l3,self.dropout_f)
      l4 = self.mlp(l3,layer_config['l3'],layer_config['l4'],name='l4')
      l4 = tf.nn.dropout(l4,self.dropout_f)
      output = self.mlp(l4,layer_config['l4'],layer_config['output'],name='output')
      return output

    def mlp(self, input_,input_dim,output_dim,name="mlp"):
      with tf.variable_scope(name):
        w = tf.get_variable('w',[input_dim,output_dim],tf.float32,tf.random_normal_initializer(mean = 0.001,stddev=0.02))            
        b = tf.get_variable('b',[output_dim],tf.float32,tf.constant_initializer(0.1))
        return tf.nn.relu(tf.matmul(input_,w)+b)
    
   
    def contrastive_loss(self):
        tmp= self.labels *tf.square(self.distance)        
        tmp2 = (1-self.labels) *tf.square(tf.maximum((1 - self.distance),0))
        return tf.reduce_sum(tmp +tmp2)/self.batch_size/2
   
    
    def set_dropout(self, dropout):
      self.dropout_f = dropout
      
    
    
# Other helper functions
def next_batch(s,e,inputs,labels):
    input1 = inputs[s:e,0]
    input2 = inputs[s:e,1]        
    y = np.reshape(labels[s:e],(len(range(s,e)),1))    
    return input1,input2,y
  
def create_pairs(x, digit_indices):
    '''Positive and negative pair creation.
    Alternates between positive and negative pairs.
    '''
    pairs = []
    labels = []
    left_digits = []
    right_digits = []
    n = min([len(digit_indices[d]) for d in range(10)]) - 1
    for d in range(10):
        for i in range(n):
            z1, z2 = digit_indices[d][i], digit_indices[d][i+1]
            pairs += [[x[z1], x[z2]]]            
            left_digits.append(d)
            right_digits.append(d)            
            inc = random.randrange(1, 10)
            dn = (d + inc) % 10
            z1, z2 = digit_indices[d][i], digit_indices[dn][i]
            pairs += [[x[z1], x[z2]]]
            labels += [1, 0]
            left_digits.append(d)
            right_digits.append(dn)
    return np.array(pairs), np.array(labels), np.array(left_digits), np.array(right_digits)

def compute_accuracy(prediction,labels):
    return labels[prediction.ravel() < 0.5].mean()
    #return tf.reduce_mean(labels[prediction.ravel() < 0.5]) 
```

```python
# Create train-test sets
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data() #get some data
train_images = train_images/255
test_images = test_images/255

print('Shapes: train_img {}, train_labels {},test_img {}, test_labels {}'.format(train_images.shape,train_labels.shape,
                                                                                  test_images.shape,test_labels.shape))

# create training+test positive and negative pairs
digit_indices = [np.where(train_labels == i)[0] for i in range(10)]
tr_pairs, tr_y, tr_left_digits, tr_right_digits = create_pairs(train_images, digit_indices)
digit_indices = [np.where(test_labels == i)[0] for i in range(10)]
te_pairs, te_y, te_left_digits, te_right_digits = create_pairs(test_images, digit_indices)

tr_pairs = tr_pairs.reshape(-1,2,tr_pairs.shape[2]*tr_pairs.shape[3])
te_pairs = te_pairs.reshape(-1,2,te_pairs.shape[2]*te_pairs.shape[3])
#print(len(left_digits), len(right_digits))
print(tr_pairs.shape, tr_y.shape)
print(te_pairs.shape, te_y.shape)
```

```python
# Launch the graph
lr = 0.001
n_epochs = 10
batch_size = 128
dropout_f = 0.5

layer_config = {'input_shape':784,'l1':200,'l2':200,'l3':100,'l4':100,'output':2}
tune_config = {'kernel_dim':5,'batch_size': batch_size}
total_batch = int(tr_pairs.shape[0]/batch_size)

tf.reset_default_graph()
with tf.Session(config=tf.ConfigProto(log_device_placement=True)) as sess:
    test_siam = siamese_net(layer_config, tune_config)
    optimizer = tf.train.AdamOptimizer(learning_rate = lr).minimize(test_siam.loss)
    #sess.run(init)
    tf.global_variables_initializer().run()

    cur_time = datetime.time(datetime.now())
    print('Start training time: {}'.format(cur_time))
    # Training cycle
    for epoch in range(n_epochs):
        avg_loss = 0.
        avg_acc = 0.

        start_time = time.time()

        # Loop over all batches
        for i in range(total_batch):
            s  = i * batch_size
            e = (i+1) *batch_size

            # Fit training using batch data
            input1,input2,y = next_batch(s,e,tr_pairs,tr_y)                        
            #print(input1.shape, input2.shape, y.shape)                

            # Train pass
            test_siam.set_dropout(dropout_f)
            _,predict,loss_value=sess.run([optimizer,test_siam.distance,test_siam.loss], 
                                          feed_dict={test_siam.input_L:input1,
                                                     test_siam.input_R:input2,
                                                     test_siam.labels:y})                

            tr_acc = compute_accuracy(predict,y)
            #print(tr_acc)

            avg_loss += loss_value
            avg_acc +=tr_acc*100

        duration = time.time() - start_time
        print('epoch %d  time: %.2f loss %0.4f acc %0.2f' %(epoch,duration,avg_loss/(total_batch),avg_acc/total_batch))

    cur_time = datetime.time(datetime.now())
    print('End training time: {}'.format(cur_time))      
    
    
    # Test model  
    test_sampx = 10000
    print('Testing model')
    print(te_pairs.shape)
    print(test_siam.dropout_f)
    test_siam.set_dropout(1)    
    print(test_siam.dropout_f)
    
    train_feature_1 = test_siam.branch_L.eval(feed_dict={test_siam.input_L:tr_pairs[:test_sampx,0]})
    train_feature_2 = test_siam.branch_R.eval(feed_dict={test_siam.input_R:tr_pairs[:test_sampx,1]})
    
    test_feature_1 = test_siam.branch_L.eval(feed_dict={test_siam.input_L:te_pairs[:,0]})
    test_feature_2 = test_siam.branch_R.eval(feed_dict={test_siam.input_R:te_pairs[:,1]})
    
    predict=test_siam.distance.eval(feed_dict={test_siam.input_L:te_pairs[:,0],test_siam.input_R:te_pairs[:,1]})
    
    y = np.reshape(te_y,(te_y.shape[0],1))
    te_acc = compute_accuracy(predict,y)
    print('Accuract test set %0.2f' % (100 * te_acc))
    
```
